# 6 . Контрольні питання.

*  Як знання закону розподілу значень пікових навантажень у комп’ютерній мережі підприємства може допомогти при моделюванні та аналізі пікових навантажень? - Знання закону розподілу пікових навантажень є фундаментальним для точного моделювання мережі, оскільки воно дозволяє реалістично імітувати рідкісні, але критичні сплески трафіку. Воно дає змогу кількісно оцінити ризики та ймовірність перевищення критичної пропускної здатності, що є основою для надійного планування. Зрештою, це знання дозволяє оптимізувати ресурси та механізми управління якістю обслуговування , забезпечуючи економічно ефективну та стабільну роботу мережі.

* Як знайти математичне сподівання функції одного випадкового аргумента? - Для ДВВ - це сума добутків значень функції на відповідні ймовірності.Для НВВ - інтеграл від добутку функції на функцію густини по всій області визначення.
  
*  Як знайти дисперсію функції одного випадкового аргумента? - Спочатку знаходимо математичне сподівання.Знаходимо математичне сподівання квадрату функції.Та рахуємо за формулою D(x)=M(x ^ 2)-M(x) ^ 2

* Чому на етапі обчислення закону розподілу функції від випадкової величини виконати аналіз монотонності функції? - Аналіз монотонності є необхідним, оскільки він дозволяє визначити, чи існує для кожного значення y лише один відповідний корінь x, що суттєво спрощує формулу перетворення розподілів. Якщо функція строго монотонна, використовується проста формула, де сума зводиться до одного доданка, а також однозначно визначається обернена функція та межі для функції. Якщо функція не монотонна, аналіз монотонності допомагає розбити область визначення на інтервали, де обернення є однозначним, що забезпечує правильність обчислень.
