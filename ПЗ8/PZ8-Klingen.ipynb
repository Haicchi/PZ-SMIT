# 2 . Контрольні питання.

* Що таке вибірковий метод і як він використовується в статистиці? - Вибірковий метод — це основний інструмент математичної статистики. Його суть полягає в тому, що замість дослідження всіх об’єктів , ми досліджуємо лише їхню частину, а висновки поширюємо на всіх.Цей метод використовується для економії часу та ресурсів, а також у випадках, коли перевірити кожен об'єкт неможливо.

* Які є основні точкові статистичні оцінки, і як вони обчислюються? - Основними точковими оцінками є вибіркове середнє, яке визначає центр розподілу даних, та дисперсія , що показує міру їхнього розсіювання.Вибіркове середнє обчислюється діленням суми всіх значень вибірки на їхню загальну кількість.Дисперсія та середнє квадратичне відхилення розраховуються на основі суми квадратів різниць між кожним значенням вибірки та її середнім арифметичним.
  
* Які фактори впливають на точність статистичних оцінок? - На точність статистичних оцінок безпосередньо впливають обсяг вибірки та мінливість самих даних у генеральній сукупності.Крім того, критично важливою є репрезентативність вибірки, яка забезпечується правильним і випадковим методом відбору об'єктів, щоб уникнути системних помилок.

* Як визначається вибіркове середнє і вибіркова дисперсія? - Вибіркове середнє обчислюється шляхом додавання всіх значень вибірки та діленням отриманої суми на загальну кількість елементів, що характеризує центр розподілу даних.Вибіркова дисперсія визначається як сума квадратів відхилень кожного значення від середнього, поділена на обсяг вибірки мінус один.

*  Що таке точні вибіркові розподілення і як вони допомагають у роботі з вибірковими оцінками? - Точні вибіркові розподілення — це теоретичні закони, що описують ймовірність появи різних значень статистик при відборі з нормальної сукупності.Вони слугують математичним еталоном, який дозволяє оцінити точність вибіркових даних та розрахувати можливу похибку репрезентативності.Саме на основі цих розподілень будуються довірчі інтервали та здійснюється перевірка статистичних гіпотез, що дає змогу робити обґрунтовані висновки про генеральну сукупність навіть за малими вибірками.

*  Які властивості мають інтервальні статистичні оцінки? - Головною властивістю інтервальної оцінки є надійність, яка визначає ймовірність того, що побудований інтервал містить істинне значення невідомого параметра генеральної сукупності.Другою властивістю є точність, яка визначається шириною цього інтервалу: чим він вужчий, тим інформативнішою є оцінка.Між ними існує компроміс: підвищення надійності неминуче призводить до розширення інтервалу, і покращити обидва показники одночасно можна лише шляхом збільшення обсягу вибірки.

* Як будується довірчий інтервал для параметра генеральної сукупності? - Спочатку обчислюється точкова оцінка параметра, яка стає центром майбутнього інтервалу.Потім, виходячи з обраного рівня надійності та мінливості даних, розраховується гранична похибка, яка визначає ширину цього інтервалу.На завершення цю похибку віднімають від центру та додають до нього, формуючи нижню та верхню межі, в яких гарантовано знаходиться шуканий параметр.
  
*  Як визначається довірчий інтервал для середнього значення генеральної сукупності? - Довірчий інтервал будується навколо знайденого вибіркового середнього шляхом додавання та віднімання від нього граничної похибки.Величина цієї похибки розраховується як добуток критичного коефіцієнта (Стьюдента для малих вибірок або Лапласа для великих) на стандартну похибку середнього.Отриманий інтервал гарантує, що з обраною ймовірністю істинне середнє значення генеральної сукупності знаходиться всередині цих меж. 

* Як використовувати інтервальні статистичні оцінки для прийняття рішень? - Інтервальні оцінки дозволяють перевірити, чи відповідає реальний процес встановленим нормативам: якщо еталонне значення не потрапляє в інтервал, процес вважається порушеним.При порівнянні двох груп рішення приймається на основі перетину їхніх інтервалів: відсутність перетину свідчить про реальну, суттєву відмінність.Також нижня та верхня межі інтервалу використовуються для оцінки ризиків, дозволяючи будувати песимістичні та оптимістичні сценарії розвитку подій. 

* Які методи можна використовувати для визначення обсягу вибірки для отримання точних статистичних оцінок? - ля визначення обсягу вибірки використовують формульний метод, де кількість спостережень розраховується на основі допустимої похибки, бажаного рівня надійності та попередньої оцінки варіації даних (дисперсії).Для якісних показників використовують формулу з максимальною дисперсією, що гарантує репрезентативність навіть за повної невизначеності.Якщо генеральна сукупність мала, отриманий результат додатково коригують спеціальним коефіцієнтом, що дозволяє зменшити необхідну кількість досліджуваних об'єктів.
