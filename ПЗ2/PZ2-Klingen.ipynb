# Мета : набути практичних навичок у розв’язанні задач з підрахунку ймовірностей на основі класичного визначення з використанням формул комбінаторики.

# 1 . В урні 4 білих та 2 чорних кульки. З цієї урни навмання взято 2 кульки. Знайти ймовірність того, що вони різного кольору. 

# 2 . У групі 30 студентів, з яких 10 відмінників. Група наугад розділений на 2 частини. Знайти ймовірність того, що в кожній частині по 5 відмінників. 

# 3 . У каталозі є 7 командних файлів і 4 текстові файли. Випадково було знищено 6 файлів. Яка ймовірність того, що було знищено 3 командні і 3 текстові файли? 

# 4 . Навмання вибирається по одній букві зі слів «дама» та «мама». Знайти ймовірність того, що ці букви: а) однакові; б) різні. 

# 5 . Навмання вибрано натуральне число, що не перевищує 20. Яка ймовірність того, що це число є дільником 20.

# 6 . Контрольні питання.

* Дати визначення класичної ймовірності. - це відношення кількості сприятливих наслідків , які сприяють настанню події A, до загальної кількості всіх рівноможливих і несумісних наслідків певного випробування.

* Що таке експеримент та простір подій у рамках теорії ймовірностей? - Це будь-яка дія, процес або спостереження, результат якого неможливо точно передбачити заздалегідь.
  
* Яким чином комбінаторика використовується для розрахунку ймовірностей за класичним методом? - Комбінаторика використовується як «механізм підрахунку» у класичній теорії ймовірностей

* У чому принципова відмінність класичного визначення ймовірності від ймовірності на просторі елементарних подій? - Принципова відмінність полягає в умовах застосування та математичній строгості визначень. Класичне визначення є спрощеним, тоді як ймовірність на просторі елементарних подій є частиною більш загальної та математично строгої аксіоматичної теорії ймовірностей.

* Наведіть інший спосіб розв’язання задачі з прикладу 2.2. - Будемо брати по одній букві на перше місце ймовірність взяти потрібну букву 1/5 і так далі до 1/1 , нам потрібні всі 5 букв тому пермножуємо ймовірності(1/5x1/4x1/3x/1/2).Результат 1/120
