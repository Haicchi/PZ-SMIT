# Мета : набути практичних навичок у розв’язанні задач з підрахунку ймовірностей на основі геометричного визначення ймовірності, алгебри подій та теорем множення і додавання ймовірностей; навчитись застосовувати на практиці формули повної ймовірності та Байєса. 

# 1 . Серед 100 лотерейних білетів є 5 виграшних. Знайти ймовірність того, що 2 наугад витягнутих білети будуть виграшними. 

# 2 . Ймовірність того, що по одному купленому білету лотереї можна виграти, складає 1/7. Знайти ймовірність того, що, купивши 5 білетів, можна: а) виграти по всім п’яти білетам; б) не виграти по жодному білету; в) виграти хоча б по одному білету.

# 3 .  Екзаменаційний білет складається з 3-х питань. Ймовірності того, що студент відповість на перше та друге питання, складають 0,9, на третє питання – 0,8. Знайти ймовірність того, що студент складе іспит, якщо для цього необхідно відповісти: а) на всі питання; б) хоча б на 2 питання. 

# 4 . Мисливець зробив три постріли по цілі, що віддаляється. Ймовірність влучення в ціль в началі стрільби складає 0,8, а після кожного пострілу зменшується на 0,1. Знайти ймовірність того, що мисливець: а) не влучить всі три рази; б) влучить хоча б один раз; в) влучить 2 рази.

# 5 . Відомо, що логін користувача комп’ютерної мережі складається з 5 маленьких латинських літер, що не повторюються, пароль складається з 6 цифр, що також не повторюються. Знайти ймовірність того, що при одній спробі можна успішно пройти авторизацію, якщо для цього необхідно правильно ввести логін і пароль.

# 6 . Контрольні питання.

* Дати визначення геометричної ймовірності - Це спосіб визначення ймовірності події у випадках, коли множина всіх можливих результатів є нескінченною, але може бути виміряна геометричними мірами.

* Навести основні правила алгебри подій - Комутативний закон , асоціативний закон , дистрибутивний закон , закон Де Моргана 
  
* Як виглядає формула множення ймовірностей для двох незалежних подій? - P(A x B) = P(A) x P(B).

* Як виглядає формула множення ймовірностей для двох залежних подій? - P(A x B) = P(A) x P(B|A).

* Як виглядає формула додавання ймовірностей для двох сумісних подій? - P(A + B) = P(A) + P(B) - P(A x B)

* Як виглядає формула додавання ймовірностей для двох несумісних подій? - P(A + B) = P(A) + P(B)

* Дати визначення повної ймовірності - P(A) = P(A ∣ H1) ⋅ P(H1) + P(A ∣ H2) ⋅ P(H2) + … + P(A ∣ Hn) ⋅ P(Hn)

* Як можна пояснити поняття апріорної та апостеріорної ймовірності, користуючись формулою Байєса? - Апріорна ймовірність - Це початкова ймовірність гіпотези , розрахована до того, як ми отримали будь-яку нову інформацію або провели випробування. Це наше попереднє переконання чи припущення, засноване на загальній статистиці, минулому досвіді чи теоретичних міркуваннях.
Апостеріорна ймовірність -  це оновлена ймовірність гіпотези (, розрахована після того, як ми отримали нове свідчення чи результат випробування (A). Вона відображає ступінь нашої впевненості у гіпотезі H, враховуючи подію A.
