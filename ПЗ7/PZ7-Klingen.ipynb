# Мета : набути практичних навичок у розв’язанні задач щодо випадкових процесів, СМО та ланцюгів Маркова. 

# 3 . Контрольні питання.

*  Що таке СМО і які основні елементи входять у її структуру? - Це Система Масового Обслуговування, математична модель, яка використовується для вивчення процесів, пов'язаних із надходженням заявок на обслуговування та їх виконанням. В її структурі виділяють три ключові елементи: вхідний потік (заявки, що надходять), обслуговуючі пристрої (канали, що виконують роботу) та чергу (місце очікування, якщо всі канали зайняті).

* Які властивості має найпростіший потік подій, і які його характеристики можна виміряти? - Найпростіший потік подій або потік Пуассона має три ключові властивості: стаціонарність (середня інтенсивність не змінюється з часом), ординарність (ймовірність одночасної появи двох чи більше подій за нескінченно малий інтервал часу є нульовою) та відсутність післядії (події незалежні одна від одної). Його головною вимірюваною характеристикою є інтенсивність потоку (середня кількість подій за одиницю часу), а також розподіл проміжків часу між подіями, який є експоненційним. 
  
* Які основні характеристики СМО визначають її продуктивність? - Основні характеристики, що визначають продуктивність СМО, включають пропускну здатність (A) — середня кількість заявок, обслужених за одиницю часу, та середню кількість зайнятих каналів (коефіцієнт використання). Важливими показниками також є ймовірність відмови (Pвідм) або абсолютні втрати, які відображають частку заявок, що не були обслужені через повну зайнятість системи.

* Які фактори впливають на інтенсивність потоку подій в системі масового обслуговування? - Інтенсивність потоку подій переважно залежить від кількості та стану джерел заявок: при обмеженій кількості джерел вона знижується, коли багато клієнтів вже перебувають у системі, оскільки менше джерел залишається активними. Додатково на інтенсивність впливає стаціонарність процесу (чи є інтенсивність сталою чи змінюється залежно від часу доби/сезону) та зовнішні чинники, такі як маркетингова активність, що може спричинити сплески надходження заявок.

*  Як визначається інтенсивність обслуговування в СМО? - Інтенсивність обслуговування, μ (мю), визначає середню кількість заявок, які один обслуговуючий канал може обробити за одиницю часу за умови, що він постійно зайнятий. Ця характеристика є оберненою до середнього часу обслуговування (1/Tобс) однієї заявки, і для моделей M/M/1/M/M/s тривалість обслуговування зазвичай має експоненційний розподіл. Таким чином, для підвищення μ необхідно зменшити середній час, потрібний для виконання однієї робочої операції.

*  Які властивості мають ланцюги Маркова, і як вони застосовуються в теорії СМО? - Ланцюги Маркова характеризуються властивістю відсутності післядії , що означає, що майбутній стан системи залежить лише від її поточного стану і не залежить від того, як вона потрапила в цей стан. У теорії СМО ланцюги Маркова застосовуються для моделювання динаміки системи, де станами є кількість заявок у черзі чи на обслуговуванні, а переходи між станами відбуваються з інтенсивностями λ та μ. Це дозволяє записати та розв'язати систему рівнянь Колмогорова для знаходження стаціонарних імовірностей Pk усіх можливих станів СМО. 

* Що таке стаціонарний режим роботи СМО і чому він важливий для аналізу? - Стаціонарний режим роботи СМО — це такий режим, коли ймовірності станів системи та її основні характеристики не залежать від часу. Це означає, що система досягла рівноваги, де середня інтенсивність надходження заявок дорівнює середній інтенсивності їх обслуговування.Він важливий для аналізу, оскільки дозволяє застосовувати алгебраїчні рівняння балансу , замість складних диференціальних рівнянь, що значно спрощує розрахунок ключових показників продуктивності СМО. Аналіз у стаціонарному режимі дає інженерам та аналітикам стабільні прогнозні значення для прийняття рішень щодо необхідної кількості ресурсів. 
  
*  Як визначається ймовірність втрати заявки в системі масового обслуговування? - Ймовірність втрати заявки визначається лише у СМО з відмовами (системах з обмеженою ємністю), де клієнт не може потрапити в систему, якщо всі місця зайняті. Вона дорівнює ймовірності того, що система перебуває у стані максимальної завантаженості — тобто коли всі обслуговуючі канали зайняті, а черга повністю заповнена. Для знаходження Pвідм необхідно спочатку розв'язати систему рівнянь Колмогорова для стаціонарного режиму, а потім взяти ймовірність останнього стану P. 

* Що таке ефективність обслуговування в СМО і як її вимірюють? - Це комплексна оцінка якості роботи системи, що відображає її здатність швидко та повноцінно обробляти вхідний потік заявок при раціональному використанні ресурсів. Вона вимірюється через середню пропускну здатність (A) — кількість фактично обслужених заявок за одиницю часу — та коефіцієнт використання обслуговуючих каналів (ρ).

* Як визначається коефіцієнт завантаження системи масового обслуговування, і чому він важливий для оцінки її продуктивності? - Коефіцієнт завантаження системи, позначається як ρ (ро), визначається як відношення інтенсивності вхідного потоку (λ) до сумарної інтенсивності обслуговування (μ) всіх каналів: ρ=λ/s⋅μ, де s — кількість каналів обслуговування.Цей показник є критично важливим для оцінки продуктивності, оскільки він показує середню частку часу, протягом якого обслуговуючі канали зайняті роботою, і визначає ступінь навантаження на систему. Чим ближче ρ до 1 (для необмеженої черги), тим гіршими стають показники якості.
